```python
import xlsxwriter as xlsx
import os

#-------------------------------------------------------caminho e  arquivo

caminho = 'C:\\Users\\FIC\\Documents\\RSC_Python\\'
arquivo = 'exercicio-6.xlsx'

#-----------------------------------------------------criar um arquivo e uma aba
workbook = xlsx.Workbook(caminho+arquivo)
sheetCustos = workbook.add_worksheet('Custos')

#---------------------------------------------------Modificando a cor de fundo das celulas
corfonte = workbook.add_format()
corfonte.set_font_color('black')


#---------Formatação celulas 
corcabecalholinhas = workbook.add_format({'fg_color':'#50AB37',# cor preenchimento
                                 'font_color':'#1D3857',# cor da fonte
                                 'bold': True, # Negrito
                                 'border':True # Aplicar bordas
                                })

corcabecalhotitulos = workbook.add_format({'fg_color':'#F39E76',# cor preenchimento
                                         'font_color':'#1D3857',# cor da fonte
                                         'bold': True, # Negrito
                                         'border': True, # Aplicar bordas
                                         'font_size':12,
                                        'align':'center',
                                           'valign':'vcenter'                                        
                                            })

#---------Formatação Valores
corvalores = workbook.add_format({'num_format':'R$#,##0'})

#---------Formatação dos totais
cortotais = workbook.add_format({'num_format':'R$#,##0',
                                'fg_color':'#F39E76',
                                'font_color':'#1D3857',
                                'bold': True, 
                                'border':True 
                                })


#formatação titulo
cortitulo = workbook.add_format({'bold':True,
                                'font_size':18,
                                'align':'center'
                                })



#------------------------------------------------------escrever nas células
sheetCustos.merge_range('A1:C1','Controle de Custos',cortitulo)
sheetCustos.merge_range('A2:A3','Itens',corcabecalhotitulos)

sheetCustos.write('A4','Salário',corcabecalholinhas)
sheetCustos.write('A5','Horas Extras',corcabecalholinhas)
sheetCustos.write('A6','Previdencia Social',corcabecalholinhas)
sheetCustos.write('A7','FGTS',corcabecalholinhas)
sheetCustos.write('A8','Aluguel',corcabecalholinhas)
sheetCustos.write('A9','Luz/Água/Telefone',corcabecalholinhas)
sheetCustos.write('A10','Mat.Expediente',corcabecalholinhas)
sheetCustos.write('A11','Outras Despesas',corcabecalholinhas)
sheetCustos.write('A12','Total',corcabecalhotitulos)

sheetCustos.merge_range('B2:C2','Total 1° Trimestre',corcabecalhotitulos)

sheetCustos.write('B3','Previsão',corcabecalhotitulos)
sheetCustos.write('B4',2000.00,corvalores)
sheetCustos.write('B5',3150.00,corvalores)
sheetCustos.write('B6',4300.00,corvalores)
sheetCustos.write('B7',5450.00,corvalores)
sheetCustos.write('B8',6600.00,corvalores)
sheetCustos.write('B9',7750.00,corvalores)
sheetCustos.write('B10',8900.00,corvalores)
sheetCustos.write('B11',10050.00,corvalores)
sheetCustos.write('B12',48200.00,cortotais)

sheetCustos.write('C3','Gastos',corcabecalhotitulos)
sheetCustos.write('C4',1550.00,corvalores)
sheetCustos.write('C5',2550.00,corvalores)
sheetCustos.write('C6',3550.00,corvalores)
sheetCustos.write('C7',4550.00,corvalores)
sheetCustos.write('C8',5550.00,corvalores)
sheetCustos.write('C9',6550.00,corvalores)
sheetCustos.write('C10',7550.00,corvalores)
sheetCustos.write('C11',8550.00,corvalores)
sheetCustos.write('C12',40400.00,cortotais)


#-----------------------------------------------------Fechar o documento e salvar
workbook.close()
#----------------------------------------------------Abrir documento
os.startfile(caminho+arquivo)
```
